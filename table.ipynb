```python
import pandas as pd
from sklearn.cluster import KMeans

df = pd.read_csv('emi.csv', low_memory = False)
df = df.fillna(0) #0 for empty cells
df['7'] = 0 #0 for projects also
df['Points'] = df[['1', '2', '3', '4', '5', '6', '7']].sum(axis = 1)
km = KMeans(n_clusters = 3, random_state = 0)
km.fit(df['Points'].values.reshape(-1, 1))
df['Results'] = km.labels_
cluster_mapping = {0: 'tochno avtomat', 1: 'vse ploho', 2: 'ne avtomat'}
df['Results'] = df['Results'].replace(cluster_mapping)
df.loc[:, ['Student', 'Points', 'Results']]
```
